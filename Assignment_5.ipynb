```python
#Q1)To find the ultimate moment carrying capacity of a singly reinforced beam
#Inputs
fck = float(input("Enter the value of characteristic compressive strength (fck): "))
fy = float(input("Enter the grade of steel / yield strength (fy): "))
Es = float(input("Enter the modulus of elasticity of steel (Es): "))
b = float(input("Enter the width of beam (b): "))
d = float(input("Enter the effective depth (d): "))
d1 = float(input("Enter the bar diameter d1: "))
d2 = float(input("Enter the bar diameter d2: "))
n = int(input("Enter the number of bars (n): "))

# 2. Area of steel
Ast1 = n * (0.7854 * d1**2)
Ast2 = n * (0.7854 * d2**2)
print("Area of steel Ast1:", Ast1)
print("Area of steel Ast2:", Ast2)

# 3. Total area of steel
Ast = Ast1 + Ast2
print("Total steel area Ast:", Ast)

# 4. Neutral axis depth factor (ku)
ku = 0.0035 / (0.0055 + (fy / (1.15 * Es)))
print("Neutral axis factor (ku):", ku)

# 5. Moment resistance factor (Ru)
Ru = 0.36 * fck * ku * (1 - 0.42 * ku)
print("Moment resistance factor (Ru):", Ru)



# 6. Maximum neutral axis depth (xumax)
xumax = ku * d
print("Maximum neutral axis (xumax):", xumax)

# 7. Actual neutral axis depth (xu)
xu = (0.87 * fy * Ast) / (0.36 * fck * b)
print("Actual neutral axis (xu):", xu)

# 8. Under/Over-reinforced check
if xu < xumax:
    print("UNDER REINFORCED")
else:
    print("OVER REINFORCED")

# 9. Use user-defined neutral axis depth (x) to compute moment of resistance (Mu)
x = float(input("Enter the neutral axis depth to use (x): "))
Mu = 0.36 * fck * x * b * (d - 0.42 * x) * 1e-6
print("Moment of Resistance (Mu):", Mu, "kN·m")
```

```python
 #Q2: Design of Slab
#Input
#Given Data
span = float(input("Enter the value of effective span in meters: "))
b1 = float(input("Enter the value of width of slab in mm: "))
bs = float(input("Enter the value of support width in meters: "))
fck = float(input("Enter the value of characteristic compressive strength (fck): "))
fy = float(input("Enter the value of grade of steel (fy): "))
Es = float(input("Enter the value of modulus of elasticity of steel (Es): "))
LL = float(input("Enter the value of live load: "))
FF = float(input("Enter the value of floor finish: "))
Density = float(input("Enter the value of density of RCC: "))

# 2. Design Constants
# Neutral axis factor
ku = 0.0035 / (0.0055 + (fy / (1.15 * Es)))
print("The value of Neutral Axis Factor (ku) is:", ku)

# Moment of resistance factor
Ru = 0.36 * fck * ku * (1 - (0.42 * ku))
print("The value of Moment Resistance Factor (Ru) is:", Ru)

# Assuming pt = 0.5 from Fig. 4 of IS 456:2000
fs = float(input("Enter the value of steel stress at service (fs): "))
MF = float(input("Enter the value of modification factor (MF): "))
S = float(input("Enter the value of span/d ratio (S): "))
k1 = float(input("Enter the value of correction factor if span > 10m (k1): "))
k2 = float(input("Enter the value of tension reinforcement correction factor (k2): "))
k3 = float(input("Enter the value of compression reinforcement correction factor (k3): "))
k4 = float(input("Enter the value of correction factor for flanged section (k4): "))

# 3. Effective depth


d1 = (span * 1000) / (S * MF * k1 * k2 * k3 * k4)
print("The value of effective depth as per deflection criteria is:", d1)

# Define effective depth and overall depth, assuming cover
d = float(input("Enter the value of effective depth in mm (d): "))
D = float(input("Enter the value of overall depth in mm (D): "))

# 4. Load Calculations
DL = D * Density / 1000
print("The dead load is:", DL)
Factor = float(input("Enter the value of partial safety factor: "))
TL = DL + LL + FF
print("The value of total load is:", TL)
Wu = Factor * TL
print("Wu =", Wu)

# 5. Bending moment (Mu)
Mu = Wu * span * span / 8
print("The value of bending moment (Mu) is:", Mu)

# 6. Effective depth check per moment
d2 = (Mu * 1e6 / (Ru * b1)) ** 0.5
print("The value of effective depth as per moment criteria is:", d2)
if d2 > d:
    print("Revise the depth: SAFE")
else:
    print("SAFE")

# 7. Minimum steel calculations
Astmin = 0.12 * b1 * D / 100
print("The value of minimum steel is:", Astmin)



# 8. Main steel calculations
Ast = (0.5 * fck * b1 * d) / fy * (1 - (1 - (4.6 * Mu * 1e6) / (fck * b1 * d * d))) ** 0.5
print("Ast:", Ast)
if Ast < Astmin:
    print("Take Ast = Astmin")
else:
    print("Ast > Astmin, hence SAFE")

# 9. Bar area and spacing calculations
dia1 = float(input("Enter the value of bar diameter for main steel: "))
dia2 = float(input("Enter the value of bar diameter for distribution steel: "))
ao1 = 0.7854 * dia1 * dia1
print("The value of area of main steel bar (ao1):", ao1)
ao2 = 0.7854 * dia2 * dia2
print("The value of area of distribution steel bar (ao2):", ao2)
Spacing1 = ao1 * b1 / Ast
print("The spacing for main steel bars is:", Spacing1)
Spacing2 = ao2 * b1 / Astmin
print("The spacing for distribution steel bars is:", Spacing2)
# Checks
print("Check 1 for main steel")
if Spacing1 > 300:
    print("UNSAFE")
else:
    print("SAFE")
print("Check 2 for main steel")
if Spacing1 > d:
    print("UNSAFE")
else:
    print("SAFE")

print("Check 1 for distribution steel")


if Spacing2 > 300:
    print("UNSAFE")
else:
    print("SAFE")
print("Check 2 for distribution steel")
if Spacing2 > 5 * d:
    print("UNSAFE")
else:
    print("SAFE")

print("Approximated values of spacing:")
S1 = float(input("Enter the value of spacing of main bars: "))
S2 = float(input("Enter the value of spacing of distribution bars: "))
Astprovided = ao1 * b1 / S1
print("The provided steel area for main bars at section in mm^2 is:", Astprovided)
Astprodist = ao2 * b1 / S2
print("The provided steel area for distribution bars at section in mm^2 is:", Astprodist)

# 10. Check for shear
Vu = (Wu * span / 2) - (Wu * ((bs / 2) - (d / 1000)))
print("The value of shear force at a section is:", Vu)
SStress = (Vu * 1000) / (b1 * d)
print("The value of shear stress is:", SStress)
SStressmax = float(input("Enter the value of maximum shear stress: "))
if SStress > SStressmax:
    print("Crushing will happen")
else:
    print("SAFE")

# 11. Percentage steel
pt = (100 * Ast) / (b1 * d)
print("The value of percentage steel is:", pt)



# 12. Shear reinforcement check
SS = float(input("Enter the value of permissible shear stress (SS): "))
k = float(input("Enter the value of depth factor k: "))
Shear = k * SS
print("The value of shear at section is:", Shear)
if SStress > Shear:
    print("Shear reinforcement required")
else:
    print("Shear reinforcement not required, SAFE")

# 13. Deflection check
ActDEF = span * 1000 / d
print("The value of span/d is:", ActDEF)
MaxDEF = S * MF * k1 * k2 * k3 * k4
print("The permissible deflection (span/d) is:", MaxDEF)
if MaxDEF > ActDEF:
    print("SAFE")
else:
    print("UNSAFE")

# 14. Anchorage length
M1 = 0.87 * fy * Ast * (d * (fy * Ast) / (fck * b1)) - 0
print("The value of Moment (M1):", M1)
lo = 8 * dia1
La = 1.3 * (M1 / Vu) + 10
print("The value of Anchorage length is:", La)

# 15. Development length
bondS = float(input("Enter the value of bond stress: "))
Ld = 0.87 * fy * dia1 / (4 * bondS * 1.6)
print("The value of development length is:", Ld)


if La > Ld:
    print("SAFE")
else:
    print("Increase anchorage")
```
